```python
import graphviz
```

```python
g = graphviz.Digraph()
```

```python
h0 = ['alg0', 'alg1', 'alg2', 'alg3']
```

```python
g.attr('node', style='filled', color='#f5f5d3')
for alg in h0:
    g.node(alg)
```

```python
g
```

```python
for i in range(len(h0)-1):
    g.edge(h0[i], h0[i+1],color='blue')
```
