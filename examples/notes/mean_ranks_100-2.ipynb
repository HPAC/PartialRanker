```python
from algorithm_ranking import MeasurementsVisualizer, MeasurementsSimulator, RankBFS, CompareAlgs
```

```python
cfg = {}
cfg['alg0']  = [0.1,0.02] #[mean, std]
cfg['alg1']  = [0.15,0.01]
cfg['alg2']  = [0.16,0.02]
cfg['alg3']  = [0.17,0.04]
cfg['alg4']  = [0.20,0.02]
h0 = list(cfg.keys())
```

```python
cfg = {}
mu1 = 0.1
mu2 = 0.12
sig = 0.02
for i in range(100):
    alg = 'alg{}'.format(i)
    if i%2:
        cfg[alg] = [mu1, sig]
    else:
        cfg[alg] = [mu2, sig]
```

```python
ms = MeasurementsSimulator(cfg, 'normal', seed=152)   
ms.measure(30,1)
data = ms.get_alg_measurements()
h0 = list(cfg.keys())
mv = MeasurementsVisualizer(data, h0)
```

```python
fig = mv.show_measurements_violinplot(h0,scale=0.2)
```

```python
compareAlgs = CompareAlgs(data, h0)
```

```python
rank = RankBFS(h0,compareAlgs)
```

```python
history, ranks = rank.rank_algorithms(q_min=25,q_max=75)
```
